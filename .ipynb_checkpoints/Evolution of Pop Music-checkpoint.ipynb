```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('input/billboardHot100_1950_2015.csv')
```

```python
# data1 = data.loc[:,['year', 'title']]
data1 = data.loc[:,['pos', 'year', 'title', 'artist',
       'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_ms', 'time_signature', 'year_bin',
       'cluster', 'Gender']]
```

```python
print("1:%d 2:%d"%(len(data1),len(data)))
```

```python
nsongs = {}
for y in range(1950,2016):
    nsongs[y] = len(data1[data1.year==y])
    print("Year=%d, nsongs=%d" % (y, nsongs[y]))
```

```python
yrs = range(1950,2016)
values = [nsongs[y] for y in yrs]
plt.bar(yrs, values, align='center')
plt.xlabel("Year")
plt.ylabel("Number of songs")
```

```python
df = data1
df_g = df.groupby('year')
```

```python
attr = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_ms', 'time_signature']
y_vals = []
i = 1
plt.figure(figsize=(15,20))
for a in attr:
    y_vals = []
    for name, groups in df_g:
    #     print(name)
        y_vals.append(groups[a].mean())
    plt.subplot(7, 2, i)
    plt.plot(yrs, y_vals)
    plt.ylabel(a)
    i += 1
```
